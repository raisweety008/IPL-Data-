```python
import pandas as pd
'''numpy is open source python library that can efficiently do numerical operations on large quantities of data.
THere are a few functions that exist in numpy that we use on pandas dataframe. Most importantly  pandas is built on top of numpy 
that is why for data analysis we import both pandas and numpy
Where pandas helps in creating structured data called as data frame and can perform data wrangling'''

import numpy as np

import warnings
warnings.filterwarnings("ignore")
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("data.csv")
df.head()
```

```python
df.info()
```

```python
#line plot
avg_runs_per_year=df.groupby("year")["runs_scored"].mean()
plt.figure(figsize=(10,6))
plt.plot(avg_runs_per_year.index,avg_runs_per_year.values,color="green",linestyle="--",marker="o")
plt.title("Average runs Scored per Year", fontsize=12)
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.show()
# insight -- This chart shows the trend of average runs scroed accross different IPL seasons. It providesa high level view of how scroing patterns
#hace changed over the yarsve.g if the line rises it suggest that matches are becoming more high scornig possibly due to improved batting echnique or
# flatter pitcher. A drop could indicate better bowling strategies or more challenging playing conditions in specific year.
```

```python
#for scatter plot u need continues values
plt.figure(figsize=(10,6))
plt.scatter(df["runs_scored"],df["wickets_taken"],color="red")
plt.title("Runs Scored vs Wickets Taken",fontsize=12)
plt.xlabel("Runs Scored")
plt.ylabel("Wickets taken")
plt.grid()
plt.show()
# Thi splot highlight sth erelationship between runs scored and wuckets taken inindividuAL MATCHES. Matches with higher runs scored
#e.d abover 200 often have fewer wickets takensuggesting 1 or 2 dominant betting performance.
# matches wit lower runs and higher wickets indicate tough balling condition. There are outliers i.e the matches with extra ordinary performances such as 
# match where runs scored exceed 250 runs. Also, this plot shows that there is no corelation between wickets taken and runs scored.
#
```

```python
#average runs scored by Teams
avg_runs_per_team=df.groupby("team")["runs_scored"].mean().sort_values(ascending=False)
plt.figure(figsize=(10,6))
plt.bar(avg_runs_per_team.index,avg_runs_per_team.values,color="purple")
plt.title("Average runs scored by Teams",fontsize=12)
plt.xlabel("Team")
plt.ylabel("Runs Scored")
plt.xticks(rotation=45)
plt.show()
#This chart shows the average runs scored by each team over the datasets period of time, teams withhigher bars e.g Rajasthan Royals and Sunrisers Hyderabad
# indicate consistant high scoring performance reflecting strong batting line ups. The lower bars e.g pUnjab Kings and Kolkata Knight riders may reflect 
#fewer seasons in IPL or weaker batting strenght
```

```python
#Histogram: Distribution of run-scred
plt.figure(figsize=(8,8))
plt.hist(df["runs_scored"],bins=15,edgecolor="black",color="yellow")
plt.title("Histogram of Runs Scored")
plt.xlabel("Runs Scored")
plt.ylabel("Frequency")
plt.show()
# This chart shows how often teams score within certain ranges. THe peak of the histogram represeants the most common range of scorest
#i.e 130 to 170 runs . THe tale on the ridgt shows occational very high scrores i.e. above 200 indicating exceptional perfomrance.
#
```

```python
# box plot: Runs scored by teams.
plt.figure(figsize=(10,7))
df.boxplot(column="runs_scored",by="team",grid=False)
plt.suptitle("") #removes by default title given by pandas)
plt.title("Boxplot : Runs Scored by Teams")
plt.xlabel("Teams")
plt.ylabel("Runs Scored")
plt.xticks(rotation=45)
plt.show()
#The box plot reveals the spread and variations in runs scored by each team . THe mediam i.e the central line in the box shows 
#the typical score for a team . The wider box or more outliars that is dots beyond wisker suggest inconssitancey in scorring while a narrow box
#implies steadyniss . The teams like Mumbai  Indians or Sunrisers Hyderabads have higher mediams refleccting consistance high performance.
```
